```python
def plot_coefficients(features, model_name):
 # 모델에서 계수 가져오기
    coefficients = model_name.coef_
    
    # 변수와 계수를 묶어줍니다.
    coef_features = list(zip(features, coefficients))
    # 계수를 기준으로 내림차순 정렬합니다.
    coef_features.sort(key=lambda x: abs(x[1]), reverse=True)
    sorted_features, sorted_coefficients = zip(*coef_features)
    print(coef_features)
    plt.figure(figsize=(10,6))
    plt.barh(sorted_features, sorted_coefficients)
    plt.xlabel('Coefficients')
    plt.ylabel('Variables')
    plt.title(f'Coefficients of Variables ({model_name})')
    plt.tight_layout()
    plt.savefig(f'{model_name}_coefficients3.png', bbox_inches='tight')
    plt.show()
```

```python
plot_coefficients(x_.columns, model1)
```

```python
def plot_variable_importance(model, feature_names, model_name):
    # 모델
    coefficients = model.coef_
    # 변수 중요도의 절대값
    importance = np.abs(coefficients)

    # 변수 중요도를 내림차순으로 정렬한 인덱스 가져오기
    sorted_indices = importance.argsort()[::-1]

    # 변수 중요도 시각화
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(sorted_indices)), importance[sorted_indices], align='center')
    plt.yticks(range(len(sorted_indices)), np.array(feature_names)[sorted_indices])
    plt.xlabel('Variable Importance')
    plt.ylabel('Variables')
    plt.title(f'Variable Importance for {model_name} Model')
    plt.tight_layout()
    plt.savefig(f'{model_name}_variable_importance3.png', bbox_inches='tight')
    plt.show()

```

```python
plot_variable_importance(feature_names = x_.columns, model = model1, model_name = 'Linear Regression')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import ttest_rel
import numpy as np

# 모델1 평가
y_pred1 = model1.predict(x_test)
errors1 = y_test - y_pred1

# 모델2 평가
y_pred2 = model2.predict(x_test)
errors2 = y_test - y_pred2

# 모델3 평가
y_pred3 = model3.predict(x_test)
errors3 = y_test - y_pred3

# 모델4 평가
y_pred4 = model4.predict(x_test)
errors4 = y_test - y_pred4

# 모델5 평가
y_pred5 = model5.predict(x_test)
errors5 = y_test - y_pred5

# 모든 모델의 예측 오차를 리스트에 저장
errors = [errors1, errors2, errors3, errors4, errors5]

# 모델 간의 p-value 계산
p_values = np.zeros((5, 5))
for i in range(5):
    for j in range(i + 1, 5):
        t_stat, p_value = ttest_rel(errors[i], errors[j])
        p_values[i, j] = p_value
        p_values[j, i] = p_value

print("P-values matrix:")
print(p_values)

# 대각선을 제외한 가장 작은 p-value 찾기
min_p_value = np.min(p_values[np.nonzero(p_values)])
min_p_value_index = np.unravel_index(np.argmin(p_values + np.eye(*p_values.shape) * np.inf), p_values.shape)

print(f"가장 작은 p-value: {min_p_value} at index {min_p_value_index}")
# p-value가 0.0788794114499071이므로, 두 모델 간의 성능 차이는 통계적으로 유의미하지 않다.
```
