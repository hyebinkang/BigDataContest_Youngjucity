```python
def plot_coefficients(features, model_name):
 # 모델에서 계수 가져오기
    coefficients = model_name.coef_
    
    # 변수와 계수를 묶어줍니다.
    coef_features = list(zip(features, coefficients))
    # 계수를 기준으로 내림차순 정렬합니다.
    coef_features.sort(key=lambda x: abs(x[1]), reverse=True)
    sorted_features, sorted_coefficients = zip(*coef_features)
    
    plt.figure(figsize=(10,6))
    plt.barh(sorted_features, sorted_coefficients)
    plt.xlabel('Coefficients')
    plt.ylabel('Variables')
    plt.title(f'Coefficients of Variables ({model_name})')
    plt.tight_layout()
    plt.savefig(f'{model_name}_coefficients.png', bbox_inches='tight')
    plt.show()
```

```python
plot_coefficients(x.columns, model5)
```

```python
def plot_feature_importances(features, model_name):
    # 모델에서 중요도 가져오기
    importances = model_name.feature_importances_
    
    # 변수와 중요도를 묶어줍니다.
    feature_importance = list(zip(features, importances))
    # 중요도를 기준으로 내림차순 정렬합니다.
    feature_importance.sort(key=lambda x: x[1], reverse=True)
    sorted_features, sorted_importances = zip(*feature_importance)
    
    plt.figure(figsize=(10,6))
    plt.barh(sorted_features, sorted_importances)
    plt.xlabel('Importance')
    plt.ylabel('Features')
    plt.title(f'Feature Importances ({model_name})')
    plt.show()
```

```python
def plot_variable_importance(model, feature_names, model_name):
    # 모델
    coefficients = model.coef_
    # 변수 중요도의 절대값
    importance = np.abs(coefficients)

    # 변수 중요도를 내림차순으로 정렬한 인덱스 가져오기
    sorted_indices = importance.argsort()[::-1]

    # 변수 중요도 시각화
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(sorted_indices)), importance[sorted_indices], align='center')
    plt.yticks(range(len(sorted_indices)), np.array(feature_names)[sorted_indices])
    plt.xlabel('Variable Importance')
    plt.ylabel('Variables')
    plt.title(f'Variable Importance for {model_name} Model')
    plt.tight_layout()
    plt.savefig(f'{model_name}_variable_importance.png', bbox_inches='tight')
    plt.show()

# 함수 호출
plot_variable_importance(model1, x.columns, 'Linear Regression')
```
