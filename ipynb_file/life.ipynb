```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('삶에_대한_만족감과_정서경험_20240417195052_영주시_KOSIS.csv', encoding='cp949')
# 첫 번째 행을 열 이름으로 설정하고 기존의 열 이름은 삭제
df1.columns = df1.iloc[0]
df1 = df1.drop(0)

# 데이터프레임의 인덱스를 재설정 (기존 인덱스는 삭제)
df1 = df1.reset_index(drop=True)
```

```python
df1
```

```python
df1['지역(시군구)에 대한 전반적 만족도'] = df1['지역(시군구)에 대한 전반적 만족도'].astype(float)
```

```python
df1 = df1[['특성별(1)','특성별(2)', '지역(시군구)에 대한 전반적 만족도']]
sex = df1.loc[df1['특성별(1)'] == '성별']
age = df1.loc[df1['특성별(1)']=='연령별']
job = df1.loc[df1['특성별(1)'] == '경제활동 상태별']
money = df1.loc[df1['특성별(1)'] == '소득수준별']
edu = df1.loc[df1['특성별(1)'] == '교육정도별']
```

```python
plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
```

```python

def draw_plot(df):
    df = df.sort_values(by='지역(시군구)에 대한 전반적 만족도')

    # 그래프 그리기
    plt.figure(figsize=(8, 6))  # 그래프의 크기 설정
    plt.bar(x = df['특성별(2)'], height= df['지역(시군구)에 대한 전반적 만족도'])

    # # y축 범위를 3으로부터 시작하도록 설정
    # plt.ylim(3, (max(age['지역(시군구)에 대한 전반적 만족도']) + 2))

    # 그래프 제목 및 축 레이블 설정
    plt.title('지역(시군구)에 대한 전반적 만족도')
    plt.xlabel('특성별')
    plt.ylabel('전반적 만족도')

    # x축 레이블이 겹치지 않도록 설정
    plt.xticks(rotation=45, ha='right')
    

    # 그래프 표시
    plt.tight_layout()  # 그래프 요소들이 겹치지 않도록 자동 조정

    plt.show()  # 그래프를 그린 후에는 반드시 닫아줘야 합니다.
```

```python
for i in [age, sex, job, money, edu]:
    draw_plot(i)
```
