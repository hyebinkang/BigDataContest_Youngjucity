```python
df1 = pd.read_csv('삶에_대한_만족감과_정서경험_20240417195052_영주시_KOSIS.csv', encoding='cp949')
df1_2 = pd.read_csv('삶에_대한_만족감과_정서경험_2021.csv', encoding='cp949')
df2 = pd.read_csv('인구증가정책_20240417195133_영주시_KOSIS.csv', encoding='cp949')
df3 = pd.read_csv('저출산_해결방안_2022.csv', encoding='cp949')
df_work = pd.read_csv('취업자_근로여건_만족도_2022.csv', encoding='cp949')
```

```python
# 첫 번째 행을 열 이름으로 설정하고 기존의 열 이름은 삭제
df1.columns = df1.iloc[0]
df1 = df1.drop(0)

# 데이터프레임의 인덱스를 재설정 (기존 인덱스는 삭제)
df1 = df1.reset_index(drop=True)
df1
```

```python
city_satisfied = df1.loc[df1['특성별(1)']=='연령별']
city_satisfied = city_satisfied.drop(columns=['특성별(1)', '삶에 대한 전반적 만족도', '어제 하루의 행복', '어제 하루의 걱정'])
city_satisfied
city_satisfied = city_satisfied.rename(columns={'특성별(2)':'연령별'})
city_satisfied
```

```python
# 첫 번째 행을 열 이름으로 설정하고 기존의 열 이름은 삭제
df2.columns = df2.iloc[0]
df2 = df2.drop(0)

# 데이터프레임의 인덱스를 재설정 (기존 인덱스는 삭제)
df2 = df2.reset_index(drop=True)
```

```python
policy = df2.loc[df2['특성별(1)'] == '연령별']
policy = policy.drop(columns=['특성별(1)', '기타'])
policy = policy.rename(columns={'특성별(2)':'연령별'})
policy
```

```python
baby = policy[['연령별', '출산 및 육아정책']]
baby
```

```python
# 첫 번째 행을 열 이름으로 설정하고 기존의 열 이름은 삭제
df3.columns = df3.iloc[0]
df3 = df3.drop(0)

# 데이터프레임의 인덱스를 재설정 (기존 인덱스는 삭제)
df3 = df3.reset_index(drop=True)
baby_policy = df3.loc[df3['특성별(1)'] == '연령별']
baby_policy
```

```python
baby_policy = baby_policy.drop(columns='특성별(1)')
baby_policy.rename(columns={'특성별(2)':'연령별'}, inplace=True)
baby_policy
```

```python
baby = pd.merge(baby, baby_policy, on='연령별')
baby = baby.set_index(baby.columns[0])
baby = baby.drop(columns='기타')
baby
```

```python
baby2 = pd.merge(city_satisfied, baby_policy, on='연령별')
baby2 = baby2.set_index(baby2.columns[0])
baby2
```

```python
target ='지역(시군구)에 대한 전반적 만족도'  ## 필요도, 출산 및 육아정책의 필요도와 가장 관련된 정책은 x 중 무엇일 것이다. 

x1 = baby2.drop(target, axis=1)
y1 = baby2[target]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_valid1 = train_test_split(x1, y1, test_size = 0.2, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(x_train1, y_train1)
y_pred2 = model2.predict(x_test1)
```

```python


# Create DataFrame
codf2 = pd.DataFrame()
codf2['feature'] = list(x1)
codf2['coef'] = model2.coef_

# Check the length of the DataFrame's index
print(len(codf2))

# Sort DataFrame
codf2.sort_values(by='coef', ascending=True, inplace=True)

# Plot
plt.barh(y=codf2['feature'], width=codf2['coef'])
plt.tight_layout()

# Save and show plot
plt.savefig('출산 정책 상관계수', dpi=100, bbox_inches='tight')
plt.show()
```
