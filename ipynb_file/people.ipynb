```python
def plot_variable_importance(model, feature_names, model_name):
    # 모델
    coefficients = model.coef_
    # 변수 중요도의 절대값
    importance = np.abs(coefficients)

    # 변수 중요도를 내림차순으로 정렬한 인덱스 가져오기
    sorted_indices = importance.argsort()[::-1]

    # 변수 중요도 시각화
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(sorted_indices)), importance[sorted_indices], align='center')
    plt.yticks(range(len(sorted_indices)), np.array(feature_names)[sorted_indices])
    plt.xlabel('Variable Importance')
    plt.ylabel('Variables')
    plt.title(f'Variable Importance for {model_name} Model')
    plt.tight_layout()
    # plt.savefig(f'{model_name}_variable_importance.png', bbox_inches='tight')
    plt.show()
```

```python
kakao = destination[['기준연월', '목적지 유형', '방문자수','전년동월방문자수', '전체 검색량', '네비게이션 전체 검색수']]
kakao
```

```python
var = pd.get_dummies(kakao['목적지 유형'], dtype=int)
df_kakao = pd.concat([kakao, var], axis=1)
df_kakao.drop(columns='목적지 유형', inplace = True)
df_kakao
```

```python
df_kakao['방문자수'] = df_kakao['방문자수'].astype(int)
df_kakao['전년동월방문자수'] = df_kakao['전년동월방문자수'].astype(int)
```

```python
target = '방문자수'

x = df_kakao.drop(['기준연월',target], axis=1)
y = df_kakao[target]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)

model_lin = LinearRegression()
model_lin.fit(x_train, y_train)
y_pred_lin = model_lin.predict(x_test)

model_dt = DecisionTreeRegressor()
model_dt.fit(x_train, y_train)
y_pred_dt = model_dt.predict(x_test)

model_rf = RandomForestRegressor()
model_rf.fit(x_train, y_train)
y_pred_rf = model_rf.predict(x_test)

model_gbr = GradientBoostingRegressor()
model_gbr.fit(x_train, y_train)
y_pred_gbr = model_gbr.predict(x_test)

```

```python
from sklearn.metrics import *
# 각 모델의 성능을 평가
print("Linear Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_lin))
print("MAE:", mean_absolute_error(y_test, y_pred_lin))
print("R-squared:", r2_score(y_test, y_pred_lin))
print()

print("Decision Tree Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_dt))
print("MAE:", mean_absolute_error(y_test, y_pred_dt))
print("R-squared:", r2_score(y_test, y_pred_dt))
print()

print("Random Forest Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
print("R-squared:", r2_score(y_test, y_pred_rf))
print()

print("Gradient Boosting Regression:")
print("MSE:", mean_squared_error(y_test, y_pred_gbr))
print("MAE:", mean_absolute_error(y_test, y_pred_gbr))
print("R-squared:", r2_score(y_test, y_pred_gbr))
print()
```

```python
# Gradient Boosting Regression 모델로 예측
y_pred = model_gbr.predict(x_test)

# 성능 평가
print("Gradient Boosting Regression 모델 성능:")
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```

```python
# 예측된 값과 실제 값이 포함된 데이터프레임 생성
df_comparison = pd.DataFrame({'실제값': y_test, '예측값': y_pred})

# 처음 몇 개의 행 출력하여 확인
print(df_comparison)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)  # 실제값과 예측값의 산점도
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')  # 대각선 라인
plt.xlabel('실제값')
plt.ylabel('예측값')
plt.title('실제값 vs. 예측값')
plt.show()
```
